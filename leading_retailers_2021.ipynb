```python
retailers['Dominant_operational_format'] = retailers['Dominant_operational_format'].str.strip()
```

```python
retailers['Dominant_operational_format'] = retailers['Dominant_operational_format'].replace({'Discount stor' : 'Discount department store'})
```

```python
#find the dominant operational format of these stores that is most popular
dominant_stores=retailers['Dominant_operational_format'].value_counts()
dominant_stores
```

```python
#create a graph of the data
plt.figure(figsize=(15,9))
dominant_stores.plot.bar(x='Dominant_operational_format', y = 'dominant_stores',title='Leading Operational Format');
```

```python
retailers['Location'] =retailers['Location'].str.strip()
```

```python
retailers['Location'] = retailers['Location'].replace({'United Statese' : 'United States'})
retailers['Location'].value_counts()
```

```python
#find the country with the most stores
countries = retailers['Location'].value_counts().nlargest(10)
countries
```

```python
#represent the dat in a graph
plt.figure
countries.plot.bar(x='Location', y='countries',color='blue')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Country')
plt.ylabel('Count')
plt.title('Top Ten Countries');
```

```python
geographies = retailers.nlargest(10,'Geographies_of_operation')[['Name_of_Company','Location','Geographies_of_operation']]
geographies
```

```python
#plot the data
sns.set_style('darkgrid')
plt.figure(figsize=(12,9))
ax = sns.barplot(x="Name_of_Company", y="Geographies_of_operation", data=geographies,dodge = False,palette='Paired', hue="Location")
ax.set(xlabel='Company', ylabel='Geographies of Operation', title='Top 10 Companies by Geographies of Operation')
plt.xticks(rotation=45, ha='right');
```

```python
net_profit_by_format = retailers.groupby('Dominant_operational_format')['FY2021_Net_profit_margin'].mean().sort_values(ascending=False)
net_profit_by_format=net_profit_by_format.reset_index()
```

```python
sns.pointplot(data=net_profit_by_format , x='Dominant_operational_format' , y='FY2021_Net_profit_margin')
plt.xticks(rotation=90)
plt.show()
```
